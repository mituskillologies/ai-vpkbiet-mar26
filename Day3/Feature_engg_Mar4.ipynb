```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
df
```

```python
df.describe()
```

```python
# separate the data => Age and EstimatedSalary
```

```python
x = df[['Age', 'EstimatedSalary']]
x
```

```python
plt.figure(figsize=(16, 9))
plt.title("age vs salary")
plt.xlabel("Age")
plt.ylabel("Estimated Salary")
plt.scatter(x.iloc[:, 0], x.iloc[:, 1], color = 'black');
plt.grid()
```

### Normalization

```python
# import the class
from sklearn.preprocessing import MinMaxScaler
```

```python
# create an object
sca = MinMaxScaler()
```

```python
# fit the data
x_scaled = sca.fit_transform(x)
```

```python
x_scaled
```

```python
pd.DataFrame(x_scaled).describe()
```

```python
plt.figure(figsize=(16, 9))
plt.title("age vs salary")
plt.xlabel("Age")
plt.ylabel("Estimated Salary")
plt.scatter(x_scaled[:, 0], x_scaled[:, 1], 
            color = 'black');
plt.grid()
```

### Standardization

```python
from sklearn.preprocessing import StandardScaler
sca = StandardScaler()
x_scaled = sca.fit_transform(x)
```

```python
pd.DataFrame(x_scaled).describe()
```

```python
plt.figure(figsize=(16, 9))
plt.title("age vs salary")
plt.xlabel("Age")
plt.ylabel("Estimated Salary")
plt.scatter(x_scaled[:, 0], x_scaled[:, 1], 
            color = 'black');
plt.grid()
```
